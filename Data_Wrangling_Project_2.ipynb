```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in bikesummary.columns:
    lr,ur=remove_outlier(bikesummary[column])
    bikesummary[column]=np.where(bikesummary[column]>ur,ur,bikesummary[column])
    bikesummary[column]=np.where(bikesummary[column]<lr,lr,bikesummary[column])
```
